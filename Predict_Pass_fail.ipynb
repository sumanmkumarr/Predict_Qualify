Import library

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

create data

```python
data = {
    "study_hour":[1,3,4,5,6,7,8,9],
    "pass_fail":[0,0,0,0,1,1,1,1]
}
```

create dataframe

```python
df = pd.DataFrame(data)
```

visualize the data

```python
plt.scatter(df['study_hour'],df['pass_fail'])
plt.xlabel("study hour")
plt.ylabel("pass or fail 1 = pass , 0 = fail")
plt.title("study hour vs pass/fail")
plt.show()
```

train the model

```python
# features (x) and labels (y)
# x - what we know- it should be 2D
x = df[['study_hour']]

# y - what we want to predict
y = df['pass_fail']

# create model
model = LinearRegression()

# train the data
model.fit(x,y)
```

predict

```python
prediction = model.predict([[10]])
print("yes" if prediction >= 0.5 else "no")
```
